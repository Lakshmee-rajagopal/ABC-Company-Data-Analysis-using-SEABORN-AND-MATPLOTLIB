```python
import pandas as pd
df = pd.read_csv("C:/Users/sheej/OneDrive/Desktop/jupyternotebook/PYTHON COMPREHENSIVE ASSESSMENT/abc_dataset.csv.csv")
df

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Correct the data in the "height" column by replacing it with random numbers between 150 and 180
df['height'] = np.random.randint(150, 181, size=len(df))
df['height']
```

```python
# ANALYSIS TASKS

# 1.1 Distribution of employees across each team
team_distribution = df['Team'].value_counts()
print(team_distribution)
```

```python
# 1.2 Calculate percentage split relative to the total number of employees

percentage_split = (team_distribution / len(df)) * 100
print(percentage_split)
```

```python
# Visualization for the above questions 1.1 and 1.2
plt.figure(figsize=(10, 6))
sns.barplot(x=percentage_split.index, y=percentage_split.values)
plt.title('Employee Distribution Across Teams')
plt.xlabel('Team')
plt.ylabel('Percentage of Employees')
plt.xticks(rotation=55)
plt.show()
```

```python
# 2. Segregate employees based on their positions
position_distribution = df['Position'].value_counts()
print(position_distribution)
```

```python
# Visualization for question 2
plt.figure(figsize=(10, 6))
position_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Position Distribution')
plt.axis('equal')
plt.show()
```

```python
# 3. Identify the predominant age group among employees.

# Define age groups
bins = [20, 30, 40, 50, 60, 70]

# Create age groups
df['age_group'] = pd.cut(df['Age'], bins=bins)

# Calculate frequency of each age group
age_group_frequency = df['age_group'].value_counts()

# Identify age group with the highest frequency
predominant_age_group = age_group_frequency.idxmax()
frequency_of_predominant_age_group = age_group_frequency.max()

print("The predominant age group among employees is:", predominant_age_group)
print("Frequency of the predominant age group:", frequency_of_predominant_age_group)
```

```python
# Visualization for question 3

import matplotlib.pyplot as plt

# Calculate frequency of each age group
age_group_frequency = df['age_group'].value_counts()

# Visualization
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=bins, color='skyblue', edgecolor='black', label='Age Distribution')
plt.axvline(x=predominant_age_group.right, color='red', linestyle='--', linewidth=2, label='Predominant Age Group')
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()

```

```python
# 4. Discover which team and position have the highest salary expenditure.

highest_salary_team = df.groupby('Team')['Salary'].sum().idxmax()
highest_salary_position = df.groupby('Position')['Salary'].sum().idxmax()

print("Team with the highest salary expenditure:", highest_salary_team)
print("Position with the highest salary expenditure:", highest_salary_position)
```

```python
# Visualization for question 4

# Convert to DataFrame or Series
highest_salary_team = df.groupby('Team')['Salary'].sum().reset_index()
highest_salary_position = df.groupby('Position')['Salary'].sum().reset_index()

# Visualization for team with the highest salary expenditure
plt.figure(figsize=(10, 6))
sns.barplot(x='Team', y='Salary', data=highest_salary_team)
plt.title('Total Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.xticks(rotation=45)
plt.show()

# Visualization for position with the highest salary expenditure
plt.figure(figsize=(10, 6))
sns.barplot(x='Position', y='Salary', data=highest_salary_position)
plt.title('Total Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary Expenditure')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Investigate if there's any correlation between age and salary, and represent it visually.

# Visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
